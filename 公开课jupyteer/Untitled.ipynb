```python
import numpy as np
import matplotlib.pyplot  as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

deta=20
x = np.arange(26-deta, 26+deta, 0.1)
y = np.arange(371-deta, 371+deta, 0.1)
X,Y=np.meshgrid(x,y)

def func(k, b):
    return 10.0*b**2 + 110.0*b*k - 10399.325389228*b + 385.0*k**2 - 61638.816655712*k + 2775200.09562443

plt.xlabel('k')
plt.ylabel('b')
Z=func(X,Y)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()




```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt

fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
X, Y = np.meshgrid(x, y) 

print(type(X),X)

Z=4*np.sin(X)+Y**2

plt.xlabel('x')
plt.ylabel('y')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```
